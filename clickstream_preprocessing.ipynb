```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv(r'podaci\clickstream+data+for+online+shopping\e-shop clothing 2008.csv', encoding='cp1252', sep=';')
print(df.columns)
```

```python
df.rename(columns={'page 1 (main category)': 'main category','page 2 (clothing model)':'clothing model'},inplace=True)
```

```python
df
```

```python
print(df.isnull().sum())
```

```python
df_encoded = df.copy()

label_cols = df_encoded.select_dtypes(include='object').columns
for col in label_cols:
    df_encoded[col] = LabelEncoder().fit_transform(df_encoded[col])


df_encoded
```

```python
df_encoded.to_csv(r'podaci\clickstream+data+for+online+shopping\e-shop clothing 2008 preprocessed.csv', index=False)
```
